```python
import pandas as pd
import os

data_dir = 'C:\\Users\\jmdv\\Desktop'
fname = os.path.join(data_dir, 'CLEANED_DATA1.csv')

data = pd.read_csv(fname)

print(data)
x = data.iloc[:, 0:2].values
y = data.iloc[:, 2].values

#-----------------Data splitting-------------------

x_train_len = len(y) - int(len(y)*(2/5))

x_train = x[0:x_train_len]
x_test = x[x_train_len:len(y)]
y_train = y[0:x_train_len]
y_test = y[x_train_len:len(y)]

#------------------Naive Bayes--------------------

from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()

gnb.fit(x_train,y_train)

gb_score = gnb.score(x_test,y_test)

print("Accuracy of GaussianNB = ",gb_score)

#-----------------Logistic Regression----------------

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression (solver = 'liblinear', multi_class = 'ovr')

lr.fit(x_train, y_train)

lr_score = lr.score(x_test, y_test)

print("Accuracy of Logistic Regression = ",lr_score)

#------------------KNN Algorithm-------------------

from sklearn.neighbors import KNeighborsClassifier

neigh = KNeighborsClassifier(n_neighbors=3)

neigh.fit(x_train,y_train)

knn_score = neigh.score(x_test,y_test)

print("Accuracy of KNN Algorithm = ", knn_score)

#----------------Label conversion to numeric values---------

for i in range(len(y)):
    if y[i] == ' "NORMAL"':
        y[i]=0
    elif y[i] == ' "ALARM"':
        y[i]=1
    
#------------------Random Forest-------------------    

from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators = 100, random_state = 0, bootstrap = False)

regressor.fit(x_train,y_train)

rf_score = regressor.score(x_test, y_test)

print("Accuracy of Random Forest Algorithm = ",rf_score)
```
